```python
import random as r


class Password:
    ''' To generate a simple password
        Class keywords:
            strength:   low-> length : 8 (By default)
                        mid-> length : 10 (By default)
                        high-> length : 12 (By default)

            length :    _int

    '''
    CHARS_PASSWD = {
        'alpha':  list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'),
        'numbers': list(str(x) for x in range(10)),
        'punctuation': list('./,*+[]{}()!@')
    }
    DIFFICULTY = {
        'low': 8, 
        'mid': 12, 
        'high': 16
    }

    @staticmethod
    def show_input_universe():
        print(Password.CHARS_PASSWD)

    def __init__(self, strength='mid', length=-1):
        self.strength = strength
        self.length = length or Password.DIFFICULTY[strength]
        self.generate_password()

    def generate_password(self):
        difficulty = list(Password.DIFFICULTY.keys())

        _characters = Password.CHARS_PASSWD['alpha'].copy()
  
        if self.strength == difficulty[1]:
            _characters += Password.CHARS_PASSWD['numbers']
  
        elif self.strength == difficulty[2]:
            _characters += Password.CHARS_PASSWD['numbers'] + Password.CHARS_PASSWD['punctuation']
  

        self.password = ''.join([r.choice(_characters) for i in range(self.length)])


```

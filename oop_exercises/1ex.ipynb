```python
import random as r


class Student:
    '''Defines a student object
    :param name: _str name of the student
    :param age:  _int. By default it's set to 10
    '''
    educational_platform = 'udemy'

    def __init__(self, name, age=10):
        self.name = name
        self.age = age

    def greet(self):
        _greeting = ['Hi, I am', 'Hey there, my name is',
                'Hi, Oh, my name is', 'Nice to meet you. I am']
        return r.choice(_greeting) + ' '+ self.name

def class_create(student_name):
    return [Student(name) for name in student_name]
```

```python
names = ['Derick','Fabricio','Ricardo','Benjamin','Lucas','Puyton']

for stud in class_create(names):
    print(stud.greet())
```
